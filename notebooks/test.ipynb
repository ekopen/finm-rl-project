```python
from experiments.common import load_spy_splits, make_single_asset_env, make_base_config, train_env_with_config
from ppo.eval_utils import run_policy_episode
from ppo.ppo_agent import PPOAgent
import numpy as np

train_df, _, test_df = load_spy_splits()
train_env = make_single_asset_env(train_df)
config = make_base_config(epochs=5)  # shorter for debug
agent = train_env_with_config(train_env, config)

test_env = make_single_asset_env(test_df)
state = test_env.reset()

actions = []
for _ in range(500):  # sample first 500 steps
    a, lp, v = agent.act(state)
    actions.append(a)
    state, r, done, info = test_env.step(a)
    if done:
        break

actions = np.array(actions)
print("Action counts:", {i: (actions == i).sum() for i in range(3)})
```
